```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
# load CSV (already uploaded in Jupyter folder)
df = pd.read_csv("ps1small.csv")
df.head()
```

```python
df["lwage"] = np.log(df["wage"])
df.head()
```

```python
df["edu_age"] = list(zip(df["education"], df["age"]))
D = pd.get_dummies(df["edu_age"], prefix="d")
```

```python
# Build clean numeric design, drop bad rows, then fit
y = pd.to_numeric(df["lwage"], errors="coerce")
X = sm.add_constant(D.drop(D.columns[0], axis=1))
Z = pd.concat([y, X], axis=1).replace([np.inf, -np.inf], np.nan).dropna()
model = sm.OLS(Z["lwage"].astype(float), Z.drop(columns="lwage").astype(float)).fit()
print(model.summary())
```
